## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2
# Classe principal
class Filme_Connie:
    def __init__Connie(self, nome, genero, duracao):
        self.nome = nome
        self.genero = genero
        self.duracao = duracao
        self.assistido = False

    def exibir_dados_Connie(self):
        status = "Assistido" if self.assistido else "N√£o assistido"
        print(f"Nome: {self.nome}")
        print(f"G√™nero: {self.genero}")
        print(f"Dura√ß√£o: {self.duracao} minutos")
        print(f"Status: {status}")
        print("-" * 30)

    def marcar_como_assistido_Connie(self):
        self.assistido = True

    @staticmethod
    def filtrar_por_genero_Connie(genero, lista_filmes):
        print(f"Filmes do g√™nero '{genero}':")
        for filme in lista_filmes:
            if filme.genero.lower() == genero.lower():
                print(f"- {filme.nome}")
        print("-" * 30)


# Subclasse herdando de Filme_Connie
class FilmeFavorito_Connie(Filme_Connie):
    def __init__Connie(self, nome, genero, duracao, nota):
        super().__init__Connie(nome, genero, duracao)
        self.nota = nota

    def avaliar_filme_Connie(self, nova_nota):
        if 0 <= nova_nota <= 10:
            self.nota = nova_nota
        else:
            print("A nota deve estar entre 0 e 10.")

    def exibir_dados_Connie(self):
        super().exibir_dados_Connie()
        print(f"Nota: {self.nota}/10")
        print("-" * 30)


# Criando objetos
filme1_Connie = Filme_Connie()
filme1_Connie.__init__Connie("A Origem", "Fic√ß√£o Cient√≠fica", 148)

filme2_Connie = Filme_Connie()
filme2_Connie.__init__Connie("O Poderoso Chef√£o", "Drama", 175)

filme3_Connie = FilmeFavorito_Connie()
filme3_Connie.__init__Connie("Interestelar", "Fic√ß√£o Cient√≠fica", 169, 9)

# Marcando um filme como assistido
filme1_Connie.marcar_como_assistido_Connie()

# Exibindo dados
filme1_Connie.exibir_dados_Connie()
filme2_Connie.exibir_dados_Connie()
filme3_Connie.exibir_dados_Connie()

# Avaliando filme favorito
filme3_Connie.avaliar_filme_Connie(10)

# Usando m√©todo de filtragem
Filme_Connie.filtrar_por_genero_Connie("Fic√ß√£o Cient√≠fica", [filme1_Connie, filme2_Connie, filme3_Connie])
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**

- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 3
# Classe principal com encapsulamento, abstra√ß√£o e sufixo _Connie
class Filme_Connie:
    def __init__Connie(self, nome, genero, duracao):
        self.nome = nome
        self.genero = genero
        self.__duracao = duracao  # Atributo privado
        self.assistido = False

    # Getter para dura√ß√£o
    def get_duracao_Connie(self):
        return self.__duracao

    # Setter para dura√ß√£o com verifica√ß√£o
    def set_duracao_Connie(self, nova_duracao):
        if nova_duracao > 0:
            self.__duracao = nova_duracao
        else:
            print("A dura√ß√£o deve ser positiva.")

    # M√©todo de l√≥gica interna (abstra√ß√£o)
    def classificar_duracao_Connie(self):
        if self.__duracao < 90:
            return "Curto"
        elif 90 <= self.__duracao <= 150:
            return "M√©dio"
        else:
            return "Longo"

    def exibir_dados_Connie(self):
        status = "Assistido" if self.assistido else "N√£o assistido"
        print(f"Nome: {self.nome}")
        print(f"G√™nero: {self.genero}")
        print(f"Dura√ß√£o: {self.__duracao} minutos ({self.classificar_duracao_Connie()})")
        print(f"Status: {status}")
        print("-" * 30)

    def marcar_como_assistido_Connie(self):
        self.assistido = True

    @staticmethod
    def filtrar_por_genero_Connie(genero, lista_filmes):
        print(f"Filmes do g√™nero '{genero}':")
        for filme in lista_filmes:
            if filme.genero.lower() == genero.lower():
                print(f"- {filme.nome}")
        print("-" * 30)


# Subclasse com heran√ßa e sufixo _Connie
class FilmeFavorito_Connie(Filme_Connie):
    def __init__Connie(self, nome, genero, duracao, nota):
        super().__init__Connie(nome, genero, duracao)
        self.nota = nota

    def avaliar_filme_Connie(self, nova_nota):
        if 0 <= nova_nota <= 10:
            self.nota = nova_nota
        else:
            print("A nota deve estar entre 0 e 10.")

    def exibir_dados_Connie(self):
        super().exibir_dados_Connie()
        print(f"Nota: {self.nota}/10")
        print("-" * 30)


# Testando o sistema com sufixo _Connie

filme1_Connie = Filme_Connie()
filme1_Connie.__init__Connie("A Origem", "Fic√ß√£o Cient√≠fica", 148)

filme2_Connie = Filme_Connie()
filme2_Connie.__init__Connie("Curtindo a Vida Adoidado", "Com√©dia", 103)

filme3_Connie = FilmeFavorito_Connie()
filme3_Connie.__init__Connie("Interestelar", "Fic√ß√£o Cient√≠fica", 169, 9)

# Marcando como assistido
filme1_Connie.marcar_como_assistido_Connie()

# Avaliando filme favorito
filme3_Connie.avaliar_filme_Connie(10)

# Exibindo dados
filme1_Connie.exibir_dados_Connie()
filme2_Connie.exibir_dados_Connie()
filme3_Connie.exibir_dados_Connie()

# Testando setter e getter
filme2_Connie.set_duracao_Connie(85)
print(f"Nova dura√ß√£o de {filme2_Connie.nome}: {filme2_Connie.get_duracao_Connie()} minutos")

# Usando filtro por g√™nero
Filme_Connie.filtrar_por_genero_Connie("Fic√ß√£o Cient√≠fica", [filme1_Connie, filme2_Connie, filme3_Connie])
```

```python
# Desenvolva aqui o c√≥digo da Parte 4
import sqlite3

# ----- CLASSES -----
class Filme_Connie:
    def __init__Connie(self, nome, genero, duracao):
        self.nome = nome
        self.genero = genero
        self.__duracao = duracao
        self.assistido = False

    def get_duracao_Connie(self):
        return self.__duracao

    def set_duracao_Connie(self, nova_duracao):
        if nova_duracao > 0:
            self.__duracao = nova_duracao

    def classificar_duracao_Connie(self):
        if self.__duracao < 90:
            return "Curto"
        elif 90 <= self.__duracao <= 150:
            return "M√©dio"
        else:
            return "Longo"

    def marcar_como_assistido_Connie(self):
        self.assistido = True

    def exibir_dados_Connie(self):
        status = "Assistido" if self.assistido else "N√£o assistido"
        print(f"[Filme] Nome: {self.nome}")
        print(f"G√™nero: {self.genero}")
        print(f"Dura√ß√£o: {self.__duracao} minutos ({self.classificar_duracao_Connie()})")
        print(f"Status: {status}")
        print("-" * 30)


class FilmeFavorito_Connie(Filme_Connie):
    def __init__Connie(self, nome, genero, duracao, nota):
        super().__init__Connie(nome, genero, duracao)
        self.nota = nota

    def avaliar_filme_Connie(self, nova_nota):
        if 0 <= nova_nota <= 10:
            self.nota = nova_nota

    def exibir_dados_Connie(self):
        super().exibir_dados_Connie()
        print(f"Nota: {self.nota}/10")
        print("-" * 30)


# ----- BANCO DE DADOS -----
def conectar_banco_Connie():
    return sqlite3.connect("catalogo_filmes.db")

def criar_tabela_Connie():
    conn = conectar_banco_Connie()
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS filmes (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT NOT NULL,
            genero TEXT NOT NULL,
            duracao INTEGER NOT NULL,
            assistido INTEGER NOT NULL,
            nota REAL
        )
    """)
    conn.commit()
    conn.close()

def inserir_filme_Connie(filme):
    conn = conectar_banco_Connie()
    cursor = conn.cursor()
    assistido = 1 if filme.assistido else 0
    nota = getattr(filme, 'nota', None)
    cursor.execute("""
        INSERT INTO filmes (nome, genero, duracao, assistido, nota)
        VALUES (?, ?, ?, ?, ?)
    """, (filme.nome, filme.genero, filme.get_duracao_Connie(), assistido, nota))
    conn.commit()
    conn.close()
    print(f"Filme '{filme.nome}' inserido no banco de dados.")

def consultar_filmes_Connie():
    conn = conectar_banco_Connie()
    cursor = conn.cursor()
    cursor.execute("SELECT id, nome, genero, duracao, assistido, nota FROM filmes")
    filmes = cursor.fetchall()
    conn.close()
    print("üé¨ Filmes no banco de dados:")
    for f in filmes:
        status = "Assistido" if f[4] else "N√£o assistido"
        nota_str = f"{f[5]}/10" if f[5] is not None else "Sem nota"
        print(f"ID: {f[0]} | Nome: {f[1]} | G√™nero: {f[2]} | Dura√ß√£o: {f[3]} min | {status} | Nota: {nota_str}")
    print("-" * 50)

def excluir_filme_Connie(filme_id):
    conn = conectar_banco_Connie()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM filmes WHERE id = ?", (filme_id,))
    conn.commit()
    conn.close()
    print(f"Filme com ID {filme_id} foi exclu√≠do.")


# ----- TESTANDO TUDO -----
if __name__ == "__main__":
    criar_tabela_Connie()

    # Criando objetos
    filme1_Connie = Filme_Connie()
    filme1_Connie.__init__Connie("Matrix", "Fic√ß√£o Cient√≠fica", 136)
    filme1_Connie.marcar_como_assistido_Connie()

    filme2_Connie = FilmeFavorito_Connie()
    filme2_Connie.__init__Connie("Gladiador", "A√ß√£o", 155, 9)
    filme2_Connie.marcar_como_assistido_Connie()

    # Inserindo no banco
    inserir_filme_Connie(filme1_Connie)
    inserir_filme_Connie(filme2_Connie)

    # Consultando dados
    consultar_filmes_Connie()

    # Exibindo via m√©todos de classe (com polimorfismo)
    print("\nüì∫ Exibi√ß√£o individual com m√©todos:")
    filme1_Connie.exibir_dados_Connie()
    filme2_Connie.exibir_dados_Connie()

    # Excluindo um filme (exemplo: ID 1)
    # excluir_filme_Connie(1)
```
